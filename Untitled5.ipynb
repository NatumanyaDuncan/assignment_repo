```python
import pandas as pd
import os
data1 = "/content/drive/MyDrive/lastweek/superhero_info - superhero_info.csv"
data2 = "/content/drive/MyDrive/lastweek/superhero_powers - superhero_powers.csv"
df1 = pd.read_csv(data1)
df1
```

```python
df2 = pd.read_csv(data2)
df2
```

```python
final_df = pd.concat([df1, df2], ignore_index=True)
final_df
```

```python
# Example cleaning operations
final_df.dropna(inplace=True)
final_df.drop_duplicates(inplace=True)


final_df['Gender'] = final_df['Gender'].astype('category')


final_df['Hero|Publisher'] = final_df['Hero|Publisher'].str.lower()

final_df['Hair color'] = final_df['Hair color'].str.lower()
final_df['Eye color'] = final_df['Eye color'].str.lower()
final_df['Skin color'] = final_df['Skin color'].str.lower()
```

```python
print(final_df.columns)
```

```python
# Assuming 'Measurements' contains both height and weight information
final_df[['Height', 'Weight']] = final_df['Measurements'].str.split('|', expand=True)

# Convert 'Height' and 'Weight' to numeric values
final_df['Height'] = final_df['Height'].str.extract(r'(\d+\.\d+)').astype(float)
final_df['Weight'] = final_df['Weight'].str.extract(r'(\d+\.\d+)').astype(float)

# Calculate average weight of heroes with Super Speed and those without
super_speed_powers = final_df[final_df['Powers'].str.contains('Super Speed', na=False)]
non_super_speed_powers = final_df[~final_df['Powers'].str.contains('Super Speed', na=False)]

avg_weight_super_speed = super_speed_powers['Weight'].mean()
avg_weight_non_super_speed = non_super_speed_powers['Weight'].mean()

print(f"Average weight of heroes with Super Speed: {avg_weight_super_speed:.2f}")
print(f"Average weight of heroes without Super Speed: {avg_weight_non_super_speed:.2f}")
```

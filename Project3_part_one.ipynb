```python
import pandas as pd
data = "/content/drive/MyDrive/queries/title-akas-us-only.csv"
df = pd.read_csv(data)
df
```

```python
# Step 1: Movies released in the United States
df = df[df['region'] == 'US']

# Step 2: Movies released between 2000 and 2021 (inclusive)
df = df[(df['titleId'].str.startswith('tt2')) | (df['titleId'].str.startswith('tt1'))]

# Step 3: Include only full-length movies
df = df[df['types'] == 'movie']

# Step 4: Exclude movies with missing values for genre or runtime
df = df.dropna(subset=['title', 'region', 'types'])

# Step 5: Include only fictional movies (exclude Documentary genre)
df = df[~df['types'].str.contains('Documentary')]


```
